# 1. Prepare dataset

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
x_data = torch.Tensor([[1.0],[2.0],[3.0]])
y_data = torch.Tensor([[2.0],[4.0],[6.0]])
```

# 2. Design model
inherit from nn.Module

```python
class LinearModel(torch.nn.Module):
    def __init__(self):
        # call father class's __init__()
        super(LinearModel,self).__init__()
        # construct an object 'linear'(in_features, out_features, bias=True)
        self.linear = torch.nn.Linear(1,1)

    def forward(self,x):
        # call callable object 'linear':__call__(self,*args, **kwargs) to forward computational graph
        y_pred = self.linear(x) # if it's for logistic regression: nn.function.sigmoid(self.linear(x)) 
        return y_pred
# instance of class LinearModel, call model(x) to compute y hat.
model = LinearModel()
```

# 3. Construct loss and optimizer
using Pytorch API
try different optimizer in linear regression:
- torch.optim.Adagrad
- Adam
- Adamx
- ASGD
- LBFGS
- RMSprop
- Rprop
- SGD

```python
# create a object(y_hat,y) to compute MSEloss
criterion = torch.nn.MSELoss(reduction='sum')   # if it's for logistic regression: BCELoss(reduction='sum')
# use model.parameters() to find all the parameters to be iterated
optimizer = torch.optim.SGD(model.parameters(), lr = 0.01)
```

# 4.Training cycle
1. forward, 
2. loss, 
3. backward, 
4. update

```python
for epoch in range(500):
    # forward: Predict
    y_pred = model(x_data)
    loss = criterion(y_pred, y_data)
    print(epoch, loss)  # loss is an object, it'll call "__str__()"
    
    optimizer.zero_grad()   # set to zero
    loss.backward()
    optimizer.step()    # update
```

# 6. Test

```python
# output weight and bias
print('w=',model.linear.weight.item())
print('b=',model.linear.bias.item())

# test model
x_test = torch.Tensor([4.0])
y_test = model(x_test)
print('y_pred = ',y_test.data)
```
