# Dataset+Dataloader
- Epoch: One forward pass and one backward pass of all the training examples.
- Batch Size: The number of training examples in one forward backward pass.
- Iteration: Nunber of passes, each pass using(batch size) nunber of examples. 

```python
import numpy as np
import torch
import matplotlib.pyplot as plt

from torchvision import datasets
from torch.utils.data import DataLoader
from torchvision import transforms  # for constructing dataloader
import torch.nn.functional as F  # for using function relu()
import torch.optim as optim     # for constructing optimizer

batch_size = 64
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307, ),(0.3081, ))
])  # convert the PIL image to tensor: R(* * *)pixle(0,1)
train_dataset = datasets.MNIST(root='./dataset/mnist/',
                            train=True,
                            download=True,
                            transform=transform)
train_loader = DataLoader(train_dataset,
                            shuffle=True,
                            batch_size=batch_size)
test_dataset = datasets.MNIST(root='./dataset/mnist/',
                            train=False,
                            download=True,
                            transform=transform)
test_loader = DataLoader(test_dataset,
                            shuffle=False,
                            batch_size=batch_size)

class Net(torch.nn.Module):
    def __init__(self):
        # call father class's __init__()
        super(Net,self).__init__()
        # construct an object 'linear'(in_features, out_features, bias=True)
        self.l1 = torch.nn.Linear(784,512)
        self.l2 = torch.nn.Linear(512,256)
        self.l3 = torch.nn.Linear(256,128)
        self.l4 = torch.nn.Linear(128,64)
        self.l5 = torch.nn.Linear(64,10)

    def forward(self,x):
        x = x.view(-1,784)
        x = F.relu(self.l1(x)) 
        x = F.relu(self.l2(x))
        x = F.relu(self.l3(x))
        x = F.relu(self.l4(x))
        return self.l5(x)
model = Net()

# create a object(y_hat,y) to compute MSEloss
criterion = torch.nn.CrossEntropyLoss()   # if it's for logistic regression: BCELoss(reduction='sum')
# use model.parameters() to find all the parameters to be iterated
optimizer = optim.SGD(model.parameters(), lr = 0.01,momentum=0.5)

def train(epoch):
    running_loss = 0.0
    for batch_idx,data in enumerate(train_loader,0):
        inputs, target = data
        optimizer.zero_grad()

        outputs = model(inputs)
        loss = criterion(outputs, target)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        if batch_idx %300 == 299:
            print('[%d, %5d] loss: %.3f' %(epoch+1, batch_idx+1, running_loss/300))
            running_loss = 0.0

def test():
    correct = 0
    total = 0
    with torch.no_grad():
        for data in test_loader:
            images, labels = data
            outputs = model(images)
            _, predicted = torch.max(outputs.data, dim=1)
            total+=labels.size(0)
            correct += (predicted == labels).sum().item()
        print('Accuracy on test set: %d %%' % (100*correct/total))

if __name__ == '__main__':
    for epoch  in range(10):
        train(epoch)
        test()
```
